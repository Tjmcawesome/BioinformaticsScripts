```python
fastqseq = open('/yourpath/3FB221_project1.fq','r')
datalist = fastqseq.readlines()

#sequence list and quality list
namelist = datalist[0::4]
seqlist = datalist[1::4]
quallist = datalist[3::4]
seqlen = len(seqlist)

#dict with sequence and name
seqplusname = dict()

avgscorelist = list()

#convert quality score ASCII to numbers
n3=0
n4=0
total=0
for x in quallist:
    for i in x:
        total += (ord(i)-32)
    avgscorelist.append(str(round((total/193),2)))
    n3 +=1
    total=0
    if n3 > len(quallist): break

#print(avgscorelist)

#sanitize names
n=0
for x in namelist:
    seqname1 = namelist[n]
    seqname2 = seqname1[1:].replace("_"," ",1)
    seqname = seqname2[:-1]
    seqplusname[seqname]=seqlist[n]
    n += 1
    if n > seqlen: break

#print dictionary
n1=0       
for x,y in seqplusname.items():
    GC=0
    GC += y.count("G")
    GC += y.count("C")
    avg = avgscorelist[n1]
    print(x,"contains percent GC = ",round((GC/193)*100,1),"% and average quality score:",avg)
    n1 += 1
    if n1 > seqlen: break
```
